```python
import matplotlib.pyplot as plt
import numpy as np

#This Algorithm is to track the changes in probability after updating based on new information. It could well be that we have a person making multiple 
#tests of a rare disease, and after each test, how certain we become that the person is actually diseased.

PH=[0.001] #Prior Probability of the Hypothesis, and each time it is updated it adds to it
PEH=0.97 #Probability of the Evidence given the Hypothesis is True
PENH=0.05 #Probability of the Evidence given the Hypothesis is not True
Num=4 #Number of successive trials with updating


for i in range(Num):
    PHE= PEH * PH[i] / ( (PEH * PH[i]) + (PENH * (1-PH[i]) ) )#Bayes Theorem
    PH.append(PHE)#Updating the prior probability



fig, graphs = plt.subplots(2, 1) #1,2 is for them to be side by side, 2,1 is for them to be underneath each other (row vector vs column vector)

graphs[0].plot(PH[0:Num],PH[1:Num+1])
graphs[0].set_title("The graph of the change in the prior probability to the posterior probability")


graphs[1].plot(PH,np.linspace(0,Num,Num+1))
graphs[1].set_title("The graph of the change in the prior probability per trial numbers")
plt.show()
```
